```python
from typing import List, Union
from purchase import Purchase
import random


class Ration:
    def __init__(self, collections:List[RecipeCollection]):
        self.collections = collections
        
    def sample(self):
        # TODO: add pop here
        random.choice(self.collections)
        
        
class RecipeCollection:
    """Class for describing collection of recipes. 
    It can also describe set of recipes for one day"""
    def __init__(self, collection:List[Union[Recipe, RecipeCollection]]):
        self.collection = collection
        self.duration = sum([it.duration for it in collection])
        
    def __iter__(self):
        for item in self.collection:
            yield item


class Recipe:
    def __init__(self, name, ingredients:List[Purchase], text, duration):
        self.text = text
        self.ingredients = ingredients
        self.full_price = sum([ingr.full_price for ingr in ingredients])
        self.full_weight = sum([ingr.full_weight for ingr in ingredients])
```
